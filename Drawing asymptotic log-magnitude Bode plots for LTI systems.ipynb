## Drawing asymptotic log-magnitude Bode plots for LTI systems
This notebook demonstrates the use of the function `bode_asymptotic` developed by the author.

#### You'd need these packages

**NumPy** (http://www.numpy.org/)

**Matplotlib** (http://matplotlib.org/), and

**Control** (https://pypi.python.org/pypi/control/0.7.0)

If you are using **Anaconda**  Python distribution (https://www.continuum.io/downloads), then you dont have to worry about NumPy and Matplotlib. You can install the Control package by running the following command in the terminal.

`pip install control`

Now it's time to import the `bode_asymptotic` package and start playing with it! 

```python
%pylab inline

from bode_asymptotic import bode_as
```

Now, say you want to draw the asymptotic log-magnitude Bode plot (hence refered to as simply *bode plot*) for the system $G(s) = \dfrac{s + 1}{s + 2}.$ Call `bode_as` with the list of coefficients.

```python
bode_as([1, 1], [1, 2])
```

The blue line represents the actual magnitude response and the green line represents the asymptotes. 

Now, if you want to draw the bode plot of $G(s) = \dfrac{s + 5}{s^2 + 3s +2}$, do this

```python
bode_as([1, 5], [1, 3, 5])
```

You can also manually declare the frequency range. Say we want the plot range of the above to be from 0.01 ras/s to 100 rad/s. Then

```python
bode_as([1, 5], [1, 3, 5], [0.01, 100])
```

If the system is given in pole-zero-gain form then you can simply pass in the poles and zeros using the numpy function `poly`. For example, to draw the Bode plot of $G(s) = \dfrac{50 (s + 2)(s + 20)}{(s+10)(s+100)}$.

```python
bode_as(50 * poly((-2, -20)), poly((-10, -100)))
```

### Notes

I haven't yet implemented the case where there are poles or zeros at the origin. So, passing such transfer functions will return an error. Other than that, everything works just fine! Systems with complex conjugate pole / zero pairs, duplicate poles / zeros would work correctly.
